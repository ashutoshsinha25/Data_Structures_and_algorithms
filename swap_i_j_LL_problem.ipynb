```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()

def swap(head, i, j):
    
    if i == j:
        return head
    
    elif i == 0 and j > 1:
        
        curr = head
        prev = None
        
        while j > 0:
            prev = curr
            curr = curr.next
            j -= 1
            
        k = curr.next
        curr.next = head.next
        prev.next = head
        head.next = k
        return curr
    
    elif i == 0 and j == 1:
        curr = head.next.next
        prev = head.next
        
        prev.next = head
        head.next = curr
        return prev
    
    
    elif j == 0 and i > 1:
        
        curr = head
        prev = None
        
        while i > 0:
            prev = curr
            curr = curr.next
            i -= 1
            
        k = curr.next
        curr.next = head.next
        prev.next = head
        head.next = k
        return curr
    
    
    elif j == 0 and i == 1:
        curr = head.next.next
        prev = head.next
        
        prev.next = head
        head.next = curr
        return prev
    
    elif i > 0 and j > 0 and abs(i-j) == 1:
        
        curr1 = head
        prev1 = head
        
        while i > 0:
            prev1 = curr1
            curr1 = curr1.next
            i -= 1
            
            
        curr2 = head
        prev2 = head
        
        while j > 0:
            prev2 = curr2
            curr2 = curr2.next
            j -= 1
            
        k = curr2.next
        prev1.next = curr2
        curr2.next = curr1
        curr1.next = k
        return head
    
    else:
        curr1 = head
        prev1 = head
        while i > 0:
            prev1 = curr1
            curr1 = curr1.next
            i -= 1
            
            
        curr2 = head
        prev2 = head
        while j > 0:
            prev2 = curr2
            curr2 = curr2.next
            j -= 1
            
        m = curr2.next
        
        prev1.next = curr2
        curr2.next = curr1.next
        prev2.next = curr1
        curr1.next = m
        return head
            
        
        
        
arr = list(int(i) for i in input().strip().split(' '))
# Create a Linked list after removing -1 from list
head = ll(arr[:-1])
i, j=list(int(i) for i in input().strip().split(' '))
head = swap(head, i, j)
printll(head)
```
