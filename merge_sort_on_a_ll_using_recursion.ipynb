```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()


def merge_func(head1, head2):
    h1 = head1
    h2 = head2
    
    finalHead = None
    finalTail = None
    
    
    if h1.data < h2.data:
        finalHead = h1
        finalTail = h1
        h1 = h1.next
        
    else:
        finalHead = h2
        finalTail = h2
        h2 = h2.next
        
    while h1 is not None and h2 is not None:
        if h1.data < h2.data:
            finalTail.next = h1
            finalTail = finalTail.next
            h1 = h1.next
        else:
            finalTail.next = h2
            finalTail = finalTail.next
            h2 = h2.next
            
    if h1 is not None:
        finalTail.next = h1
    else:
        finalTail.next = h2
    return finalHead


def mergeSort(head):
    
    if head == None:
        return
    
    if head.next == None:
        return head
    
    
    slow = head
    fast = head
    
    while fast.next is not None and fast.next.next is not None:
        
        slow = slow.next
        fast = fast.next.next
        
    head2 = slow.next
    slow.next = None
    
    x = mergeSort(head)
    y = mergeSort(head2)
    
    z = merge_func(x, y)
    return z
    
    
arr =list(int(i) for i in input().strip().split(' '))
head = ll(arr[:-1])
k = mergeSort(head)
printll(k)
```
