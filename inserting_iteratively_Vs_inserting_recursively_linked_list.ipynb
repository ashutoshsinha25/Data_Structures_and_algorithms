```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head


#usually its O(i)
#but if n >= it will be O(n)
def insertRec(head, i, data):
    
    if i == 0:
        newNode = Node(data)
        newNode.next = head
        return newNode
    
    if head is None or i < 0:
        return head
    
    
    smallHead = insertRec(head.next, i-1, data)
    head.next = smallHead
    return head



#this is O(n)
def insertIth(head, data, i):
    
    newNode = Node(data)
    if i == 0:
        newNode.next = head
        return newNode
    else:
        
        curNode = head
        count = 0
        
        while curNode is not None and i >= 0:
            if i == 0:           
                prev.next = newNode
                newNode.next = curNode
                return head
             
            prev = curNode
            curNode = curNode.next
            i -= 1
            
            if curNode is None and i == 0:
                prev.next = newNode
                return head
            
        if i != 0:
            return head
```
