```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()


def reverse(head):
    
    if head == None:
        return 
    
    if head.next == None:
        return head
    
    smallHead = reverse(head.next)
    
    s = head
    head = head.next
    head.next = s
    s.next = None

    return smallHead


def kReverse(head, k):
    
    if head == None:
        return None
    
    if head.next == None:
        return head, head
    
    count = 1
    head1 = head
    tail1 = head
    
    while count < k and tail1 is not None:
        tail1 = tail1.next
        count += 1
        
        
    if tail1 is not None:
        head2 = tail1.next
        tail1.next = None
        newTail = head1
        
        revHead = reverse(head1)
        
        head1 = revHead
        tail1 = newTail
        
        if head2 is None:
            return head1, tail1
        
        else:
            smallHead, smallTail = kReverse(head2, k)
            tail1.next = smallHead
            return head1, smallTail
        
    else:
        newTail = head1
        revHead = reverse(head1)
        return revHead, newTail
    
arr=list(int(i) for i in input().strip().split(' '))
head = ll(arr[:-1])
i=int(input())
head, tail = kReverse(head, i)
printll(head)
```
