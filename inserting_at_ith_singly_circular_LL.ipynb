```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
        
    last.next = head
    return head

def printll(head):
    temp = head
    while temp.next is not head:
        print(temp.data, end=' ')
        temp = temp.next
    print(temp.data, end = " ")

    
def lenght(head):
    temp = head
    count = 0
    while temp.next is not head:
        count += 1
        temp = temp.next
    return count
    
    
    
def insertNode(head, data, i):
    if head is not None:
    
        k = lenght(head)
        if k > 0:

            if i == 0:
                temp = head

                while temp.next is not head:           

                    temp = temp.next

                    if temp.next is head:
                        newNode = Node(data)
                        temp.next = newNode
                        newNode.next = head
                        return newNode

            else:
                if i <= k:
                    prev = None
                    curr = head
                    while i >= 0:

                        prev = curr
                        curr = curr.next
                        i -=1

                        if i == 0:
                            newNode = Node(data)
                            prev.next = newNode
                            newNode.next = curr
                            return head


                elif i == k+ 1:
                    curr = head

                    while curr.next is not head:

                        curr = curr.next

                        if curr.next is head:
                            last = curr
                            newNode = Node(data)
                            curr.next = newNode
                            newNode.next = head
                            return head

                else:
                    return head
                
        else:
            if k == 0:
                if i == 0:
                    newNode = Node(data)
                    newNode.next = head
                    head.next = newNode
                    return newNode
                    
                elif i == 1:
                    newNode = Node(data)
                    newNode.next = head
                    head.next = newNode
                    return head
                
                else:
                    return head
```
