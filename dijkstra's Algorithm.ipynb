```python
import queue
import sys
class Graph:
    def __init__(self, nVerticies):
        self.nVerticies = nVerticies
        self.adjMatrix = [ [0 for i in range(nVerticies)] for j in range(nVerticies)]
        
        
    def addEdge(self, v1, v2, wt):
        self.adjMatrix[v1][v2] = wt
        self.adjMatrix[v2][v1] = wt
        
    def removeEdge(self, v1, v2):
        if self.containsEdge(v1, v2) is False:
            return
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    
    def containsEdge(self, v1, v2):
        return True if self.adjMatrix[v1][v2] > 0 else False
    
    
    def __str__(self):
        return str(self.adjMatrix)
    

    def bfs_helper(self,sv,visited):
        q=queue.Queue()
        q.put(sv)
        visited[sv]=True
        while not q.empty():
            v=q.get()
            print(v,end=" ")
            for i in range(self.nVerticies):
                if self.adjMatrix[v][i]>0 and visited[i] is False:
                    q.put(i)
                    visited[i]=True
    def bfs(self):
        visited=[False for i in range(self.nVerticies)]
        for i in range(self.nVerticies):
            if visited[i] is False:
                self.bfs_helper(i,visited)
                
    
    def __get_minVertex(self, visited, distance):
        
        minVertex = sys.maxsize
        for i in range(self.nVerticies):
            if visited[i] is False and distance[i] < minVertex:
                minVertex = distance[i]
                min_index = i
                
        return min_index
    
                
    def djikstra(self):
        visited = [False for i in range(self.nVerticies)]
        distance = [sys.maxsize for i in range(self.nVerticies)]
        distance[0] = 0
        
        for i in range(self.nVerticies -1):
            minVertex = self.__get_minVertex(visited, distance)
            visited[minVertex] = True
            
            
            #explore the neighbours of min Vertex which are
            #not visited and update thier distances if the need arises.
            
            for j in range(self.nVerticies):
                if self.adjMatrix[minVertex][j] > 0 and visited[j] is False:
                    if distance[j] > distance[minVertex] +  self.adjMatrix[minVertex][j]:
                        distance[j] = distance[minVertex] +  self.adjMatrix[minVertex][j]
                        
        for i in range(self.nVerticies):
            print(str(i) + " " + str(distance[i]))
```

```python
li = [int(ele) for ele in input().split()]
V = li[0]
E = li[1]
g = Graph(V)

for i in range(E):
    curr_input = [int(ele) for ele in input().split()]
    src = curr_input[0]
    dest = curr_input[1]
    wt = curr_input[2]
    g.addEdge(src, dest, wt)
    
g.djikstra()
```
